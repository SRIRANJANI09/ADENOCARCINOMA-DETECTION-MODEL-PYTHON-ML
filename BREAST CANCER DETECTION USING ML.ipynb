# Importing libraries

importing and loading the necessary python libraries and the breast cancer dataset provided by Scikit-learn

implement a Naive Bayes algorithm in Machine Learning using Python.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

# Loading the Dataset

```python
data = load_breast_cancer()
```

# Extracting Features and Labels

```python
label_names = data["target_names"]
labels = data["target"]
feature_names = data["feature_names"]
features = data["data"]
```

To better understand our dataset, let’s take a look at our data by printing our class labels, the label for the first data instance, our entity names, and the entity values for the first data instance

# Viewing the Data

```python
print(label_names)
print("Class label :", labels[0])
print(feature_names)
print(features[0], "\n")
```

# Splitting The Dataset

```python
train, test, train_labels, test_labels = train_test_split(features, labels,
                                                          test_size=0.2,
                                                          random_state=42)
```

# Training the Naive Bayes Model

```python
gnb = GaussianNB()
gnb.fit(train, train_labels)
```

# Making Predictions

```python
preds = gnb.predict(test)
print(preds, "\n")
```

Using the array of true class labels, we can assess the accuracy of our model’s predictors by comparing the two arrays (test_labels vs preds).

I’ll use the accuracy_score () function provided by Scikit-Learn to determine the accuracy rate of our machine learning classifier:

# Evaluating Model Performance

```python
print(accuracy_score(test_labels, preds))
```

As you can see from the output above, our breast cancer detection model gives an accuracy rate of almost 97%. This means that 97% of the time the classifier is able to make the correct prediction.
